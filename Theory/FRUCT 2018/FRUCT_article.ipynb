```python
import os
import scipy.signal
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
DATA_FOLDER = 'C:/_Repositories/Satellite_Segmentation'
```

```python
epochs = np.loadtxt(os.path.join(DATA_FOLDER, 'epochs.txt'))
```

```python
unet = np.loadtxt(os.path.join(DATA_FOLDER, 'unet.txt'))
```

```python
linknet = np.loadtxt(os.path.join(DATA_FOLDER, 'linknet_smooth.txt'))
```

```python
segnet = np.loadtxt(os.path.join(DATA_FOLDER, 'segnet.txt'))
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
mpl.rcParams['font.size'] = 20
mpl.rc('lines', linewidth=3)
plt.figure(figsize=(20, 12))
plt.plot(epochs, smooth(unet, 1), label='UNet')
plt.plot(epochs, smooth(linknet, 1), label='TLinkNet')
plt.plot(epochs, smooth(segnet, 2), label='SegNet')
plt.xlabel('E')
plt.ylabel('DSC')
plt.legend()
plt.savefig('val.png')
```
